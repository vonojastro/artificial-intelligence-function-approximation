# **Question A: What is function approximation? - theory**

**Function Approximation**

• Function approximation is a mathematical concept on which it tries to solve a target function's result by estimating and adjusting its own approximation.

• Function approximation uses a trial and error process by which it adjusts and estimates the parameters of its data iteratively until it arrives at a more accurate value to its target result.

• It is the process of estimating a simpler function that results closely to the behavior of the original function.

• Function approximation methods aim to find a suitable representation of the original function that can be easily manipulated and analyzed for predictive purposes, making it a valuable tool where precise mathematical descriptions used in models are too complex or hidden.

# **Question B: Role of neural network in function approximation -theory**

**Neural Network in Function Approximation**

Neural networks take advantage of their model in learning the target functions behavior by adjusting its weights, biases and neuron activations in order to come up to a value which is somehow near the target functions value.

Activation functions and perceptrons are used in Neural Networks to approximate the target function. Activation functions are an important element of neural networks because they support the network to model complex nonlinear relationships between the input and output data. Activation functions and perceptrons enable neural networks to approximate non-linear relationships, this makes the Neural Networks more versatile in creating solutions for pattern recognition and value predictions.

# **Question C: Develop a python program to show how neural networks are applied in functionapproximation. –python notebook**

**Choosing a non-linear function to approximate.**

# *y(x) = sin(5π * x) + sin(6π * x)*

**Import numpy package and define a python function based on the non-linear function mentioned above.**

```python
import numpy as np

def y(x):
    return (np.sin(5*np.pi*x) + np.sin(6*np.pi*x)) # non-linear sine function
```

**Choose a set values for '*x*' from -1 to 1 then apply the non-linear function to get the value for '*y*'**

```python
x_vals = np.arange(-1,1,0.002)
y_vals = y(x_vals)
```

**Normalize y value to 0-1**

```python
y_max = y_vals.max()
y_vals /= y_max
```

**Import train_test_split package and split the data set with a training size of 80% and test size of 20%**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_vals, y_vals, test_size=0.20)
```

**Import matplotlib pyplot package and plot the generated y values to the graph, blue dots will be our training data and yellow x's will be our target values.**

**The ANN model should somehow approximate and plot its estimated values like the graph below.**

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(x_train, y_train, 'o')
plt.plot(x_test, y_test, 'x')
plt.show()
```

**Reshape the x value so that the network can understand it and easily use it for training**

```python
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
```

**Import MLPRegressor and create an ANN model with 65 input layer, 80x80x80 (80 neurons x 3 layers) hidden layer and 86 output layer with 600 iterations.**

**Train the model with training dataset and make predictions.**

**ReLU is the default activation function for the MLP Neural Network Model**

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(
    hidden_layer_sizes= [65,80,80,80,86],
    max_iter=6000,
)

mlp.fit(x_train,y_train)

predictions = mlp.predict(x_test)
```

**Plot the predictions and compare the result to the actual result of the function.**

```python
#plt.plot(x_train, y_train, 'o')
plt.plot(x_test, predictions, 'ro')
plt.show()
```

**Another MLPRegressor model with 65 input layer, 40x40x40x40x40x40 (40 neurons x 6 layers) hidden layer and 86 output layer with 600 iterations.**

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(
    hidden_layer_sizes= [65,40,40,40,40,40,40,86], #[30,80,80,80,80,80,80], #change number of layers
    max_iter=6000,
)

mlp.fit(x_train,y_train)

predictions = mlp.predict(x_test)

#plt.plot(x_train, y_train, 'o')
plt.plot(x_test, predictions, 'ro')
plt.show()
```
